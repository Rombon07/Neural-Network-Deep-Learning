```python
import numpy as np

# Data
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
t = np.array([0, 0, 0, 1])

class Perceptron:
    def __init__(self, nin): 
        rand = np.random.RandomState(1) 
        self.w = rand.normal(loc=0.0, scale=0.01, size=nin)  
        self.b = np.float_(0.0)  # b = w0
    
    def fit(self, X, y, epochs=5, eta=0.01): 
        for _ in range(epochs): 
            for i in range(X.shape[0]): 
                z = np.dot(X[i], self.w) + self.b 
                yout = np.where(z >= 0.0, 1, 0) 
                if yout != y[i]: 
                    delta = eta * (y[i] - yout) 
                    self.w += delta * X[i]  
                    self.b += delta
                    
            print(f"Weights after training (W): {self.w}")
            print(f"Bias after training (b): {self.b}\n")
    
    
    def predict(self, x): 
        z = np.dot(x, self.w) + self.b 
        return np.where(z >= 0.0, 1, 0)

# Instantiate the perceptron
NN = Perceptron(X.shape[1])

print("[INFO] training perceptron...")
NN.fit(X, t, epochs=5)

print("[INFO] testing perceptron...")
for (x, target) in zip(X, t): 
    pred = NN.predict(x)
    print("[INFO] data={}, ground-truth={}, pred={}".format(x, target, pred)) 
```

```python
import matplotlib.pyplot as plt  

# Scatter plot dengan decision boundary for i, label in enumerate(t):
for i, label in enumerate(t):
    if label == 0:
        plt.scatter(X[i][0], X[i][1], color='red', label='0' if i == 0 else "")
    else:
        plt.scatter(X[i][0], X[i][1], color='blue', label='1' if i == 3 else "")

#Menghitung decision boundary
x1_vals = np.array([0, 1])
x2_vals = -(NN.w[0] * x1_vals + NN.b) / NN.w[1]

#Plot decision boundary
plt.plot(x1_vals, x2_vals, color='green', label='Decision Boundary')

plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Perceptron Decision Boundary')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for AND logic gate
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])  # Input data
t = np.array([0, 0, 0, 1])  # Target output (AND gate)

# Define a simple Perceptron class
class SimplePerceptron:
    def __init__(self, input_size):
        self.w = np.array([0.5, -0.5])  # Initial weights (corrected to self.w)
        self.b = 0.0  # Bias term

    def fit(self, X, y, epochs=10, learning_rate=0.1):
        for _ in range(epochs):
            for i in range(X.shape[0]):
                z = np.dot(X[i], self.w) + self.b  # Corrected self.W to self.w
                y_pred = 1 if z >= 0 else 0
                # Update weights and bias using Perceptron learning rule
                self.w += learning_rate * (y[i] - y_pred) * X[i]
                self.b += learning_rate * (y[i] - y_pred)
            
            print(f"Weights after training (W): {self.w}")
            print(f"Bias after training (b): {self.b}\n")
                
                
    def predict(self, x):
        z = np.dot(x, self.w) + self.b  # Corrected self.W to self.w
        return 1 if z >= 0 else 0


# Create and train the perceptron
perceptron = SimplePerceptron(X.shape[1])
perceptron.fit(X, t)

# Test the perceptron
for x, target in zip(X, t):
    prediction = perceptron.predict(x)
    print(f"Input: {x}, Target: {target}, Prediction: {prediction}")

# Scatter plot with decision boundary
for i, label in enumerate(t):
    if label == 0:
        plt.scatter(X[i][0], X[i][1], color='red', label='0' if i == 0 else "")
    else:
        plt.scatter(X[i][0], X[i][1], color='blue', label='1' if i == 3 else "")

# Calculate decision boundary
x1_vals = np.array([0, 1])
x2_vals = -(perceptron.w[0] * x1_vals + perceptron.b) / perceptron.w[1]

# Plot decision boundary
plt.plot(x1_vals, x2_vals, color='green', label='Decision Boundary')

plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Perceptron Decision Boundary - AND Logic Gate')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data untuk gerbang logika AND
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])  # Data input
t = np.array([0, 0, 0, 1])  # Target output (gerbang AND)

# Definisikan kelas Perceptron sederhana
class SimplePerceptron:
    def __init__(self, input_size):
        self.w = np.array([0.5, -0.5])  # Bobot awal
        self.b = 0.0  # Bias

    def fit(self, X, y, epochs=10, learning_rate=0.1):
        for _ in range(epochs):
            for i in range(X.shape[0]):
                z = np.dot(X[i], self.w) + self.b  # Hitung output
                y_pred = 1 if z >= 0 else 0
                # Update bobot dan bias
                self.w += learning_rate * (y[i] - y_pred) * X[i]
                self.b += learning_rate * (y[i] - y_pred)

            print(f"Weights after training (W): {self.w}")
            print(f"Bias after training (b): {self.b}\n")
                
    def predict(self, x):
        z = np.dot(x, self.w) + self.b  # Hitung output untuk prediksi
        return 1 if z >= 0 else 0

# Buat dan latih perceptron
perceptron = SimplePerceptron(X.shape[1])
perceptron.fit(X, t)

# Uji perceptron
for x, target in zip(X, t):
    prediction = perceptron.predict(x)
    print(f"Input: {x}, Target: {target}, Prediction: {prediction}")

# Scatter plot dengan boundary keputusan
for i, label in enumerate(t):
    if label == 0:
        plt.scatter(X[i][0], X[i][1], color='red', label='0' if i == 0 else "")
    else:
        plt.scatter(X[i][0], X[i][1], color='blue', label='1' if i == 3 else "")

# Hitung boundary keputusan
x1_vals = np.array([0, 1])
x2_vals = -(perceptron.w[0] * x1_vals + perceptron.b) / perceptron.w[1]

# Plot boundary keputusan
plt.plot(x1_vals, x2_vals, color='green', label='Decision Boundary')

plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Perceptron Decision Boundary - AND Logic Gate')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
